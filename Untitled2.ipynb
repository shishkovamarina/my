```python
# для анализа
import numpy as np
import statsmodels.api as sm
%pylab inline
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import warnings
from itertools import product
from statsmodels.tsa.seasonal import seasonal_decompose

def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))

# для визуализации
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
#загрузка data
df = pd.read_excel('C:/Users/User/Desktop/MyProject/TTF quotes.xlsx', index_col=[0],parse_dates=[0])
df
```

```python
#визуализация
plt.figure(figsize(12,7))
df.quotation.plot()
pylab.show()
```

```python
#видим, что есть нулевое значение, проведем интерполяцию пропущенных значений
df.loc[df['quotation'] == 0, 'quotation'] = 'Nan'
```

```python
df['quotation'] = df['quotation'].interpolate()
df=df.astype(float)
plt.figure(figsize(12,7))
df.quotation.plot()
pylab.show()
```

```python
# группируем по дням на всякий случай
df = df.groupby(pd.Grouper(freq='1D')).aggregate("mean")
# приводим к одному виду
df.index.rename("month", inplace=True);
gas = df.iloc[:,0].rename("quotation");
df = pd.DataFrame([gas]).T
```

```python
df['quotation'] = df['quotation'].interpolate()
```

```python
df
```

Проверка стационарности и STL-декомпозиция ряда:

```python
plt.figure(figsize(12,7))
sm.tsa.seasonal_decompose(df.quotation).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df.quotation)[1])
```

Стабилизация дисперсии

Сделаем преобразование Бокса-Кокса для стабилизации дисперсии:

```python
df['quotation_box'], lmbda = stats.boxcox(df.quotation)
plt.figure(figsize(12,7))
df.quotation_box.plot()
plt.ylabel(u'Transformed wine sales')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df.quotation_box)[1])
```

Стационарность

Критерий Дики-Фуллера отвергает гипотезу нестационарности, но визуально в данных виден тренд. Попробуем сезонное дифференцирование; сделаем на продифференцированном ряде STL-декомпозицию и проверим стационарность:

```python
df['quotation_box_diff'] = df.quotation_box - df.quotation_box.shift(12)
plt.figure(figsize(12,10))
sm.tsa.seasonal_decompose(df.quotation_box_diff[12:]).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df.quotation_box_diff[12:])[1])
```

Критерий Дики-Фуллера не отвергает гипотезу нестационарности, и полностью избавиться от тренда не удалось. Попробуем добавить ещё обычное дифференцирование:

```python
df['quotation_box_diff2'] = df.quotation_box_diff - df.quotation_box_diff.shift(1)
plt.figure(figsize(12,10))
sm.tsa.seasonal_decompose(df.quotation_box_diff2[13:]).plot()   
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df.quotation_box_diff2[13:])[1])
```

Гипотеза нестационарности отвергается, и визуально ряд выглядит лучше — тренда больше нет.

Подбор модели

Посмотрим на ACF и PACF полученного ряда:

```python
plt.figure(figsize(12,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df.quotation_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df.quotation_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

```python
ps = range(0, 5)
d=1
qs = range(0, 3)
Ps = range(0, 2)
D=1
Qs = range(0, 2)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(df.quotation_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Лучшая модель:

```python
print(best_model.summary())
```

Её остатки:

```python
plt.figure(figsize(12,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

```python
df['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(12,7))
df.quotation.plot()
df.model[13:].plot(color='r')
plt.ylabel('Wine sales')
pylab.show()
```
